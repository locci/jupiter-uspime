## Abstract
Abstract of the dataset (describe the dataset with your own words):

Em Portugal, o ensino médio é composto por 3 anos e precedido por 9 anos de educação básica. Como em muitos países europeus o método de pontuação “20-point grading scale ” é adotado, onde 0 é a menor nota possível e 20 a maior nota possível. Durante o ano escolar, os alunos são avaliados três vezes, obtendo as notas G1, G2 e G3 consecutivamente.  A ultima  avaliação corresponde a nota final.
O dataset é composto por dados coletados entre o ano escolar de 2005-2006 em duas escolas publicas da região de Alentejo em Portugal. A base de dados foi criada com as informações vindas de duas fontes diferentes. Relatórios escolares em papel, contendo alguns atributos como as notas dos três períodos e o número de faltas de cada aluno. A segunda é um questionário eletrônico, contendo perguntas de âmbito demográfico, social e emocional
que esperasse afetar a desempenho dos estudantes.Todas as respostas para as perguntas contidas no questionário eletrônico são fechadas (opções pré-definidas). Dois datasets foram criados utilizando as duas fontes de dados, um relacionado as notas de Matemática e outro as notas de português.

Os Fatores Domésticos são relacionados a atributos familiares (Ex: trabalho da mãe, trabalho do pai, tamanho da familia e etc ) e informações do ambiente residencial do estudante (Ex: localização da residência, urbana ou rural e etc).
