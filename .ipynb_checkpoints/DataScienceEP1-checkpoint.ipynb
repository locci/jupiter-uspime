## Abstract
Abstract of the dataset (describe the dataset with your own words):

Em Portugal, o ensino médio é composto por 3 anos e precedido por 9 anos de educação básica. Como em muitos países europeus o método de pontuação “20-point grading scale ” é adotado, onde 0 é a menor nota possível e 20 a maior nota possível. Durante o ano escolar, os alunos são avaliados três vezes, obtendo as notas G1, G2 e G3 consecutivamente.  A ultima  avaliação corresponde a nota final.
O dataset é composto por dados coletados entre o ano escolar de 2005-2006 em duas escolas publicas da região de Alentejo em Portugal. A base de dados foi criada com as informações vindas de duas fontes diferentes. Relatórios escolares em papel, contendo alguns atributos como as notas dos três períodos e o número de faltas de cada aluno. A segunda é um questionário eletrônico, contendo perguntas de âmbito demográfico, social e emocional
que esperasse afetar a desempenho dos estudantes.Todas as respostas para as perguntas contidas no questionário eletrônico são fechadas (opções pré-definidas). Dois datasets foram criados utilizando as duas fontes de dados, um relacionado as notas de Matemática e outro as notas de português.

Descrição do Experimento

Inicialmente tomamos os dois bancos de dados fornecidos pelos autores da pesquisa original, Using data mining to predict secondary school student perfromance, student-mat.csv e student-por.csv e criamos, por meio da fusão destes dois arquivos, um terceiro arquivo denominado denominado merge-student.csv. Possuem em média 4 horas livres após a aula em um máximo de 5 horas livres identificado na amostra (fig. 29, fig. 44),  costumam passar em média 3 horas com amigos (fig. 30, fig. 44).  O desempenho nas disciplinas de matemática e português foi marcado por uma maioria dos estudantes que foi aprovada em ambas as matérias  e uma minoria de estudantes que reprovou em ambas (fig. 38). A mesma análise levando em conta a situação marital dos pais demonstra que há uma tendência para que pais que vivam juntos tenham filhos que reprovam em matemática e passem em português  ou reprovam em ambas (fig. 39).
```python
#########################################################################
#prop. Alexandre                                                        #
#inic. 6/9/2018                                                         #
#finalidade: curso de doutorado                                         #
#baseado em https://mubaris.com/2017/10/01/kmeans-clustering-in-python/ #
#########################################################################

from copy import deepcopy
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import LabelEncoder

plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')


#data = pd.read_csv('merge-student.csv')
data = pd.read_csv('merge-student-clear.csv')

for column in data.columns:
    if data[column].dtype == type(object):
        le = LabelEncoder()
        data[column] = le.fit_transform(data[column])

f1 = data['G3x'].values
f2 = data['G3y'].values
X = np.array(list(zip(f1, f2)))

#plt.scatter(f1, f2, c='black', s=7)

# Euclidean Distance Caculator
def dist(a, b, ax=1):
    return np.linalg.norm(a - b, axis=ax)

# Number of clusters
k = 4
# X coordinates of random centroids
C_x = np.random.randint(0, np.max(X), size=k)
# Y coordinates of random centroids
C_y = np.random.randint(0, np.max(X), size=k)
#C = np.array(list(zip(C_x, C_y)), dtype=np.float32)
C = np.array(list(zip(C_x, C_y)))

# Plotting along with the Centroids
#plt.scatter(f1, f2, c='#050505', s=7)
#plt.scatter(C_x, C_y, marker='*', s=200, c='g')

# To store the value of centroids when it updates
C_old = np.zeros(C.shape)
# Cluster Lables(0, 1, 2)
clusters = np.zeros(len(X))
# Error func. - Distance between new centroids and old centroids
error = dist(C, C_old, None)
# Loop will run till the error becomes zero
while error != 0:
    # Assigning each value to its closest cluster
    for i in range(len(X)):
        distances = dist(X[i], C)
        cluster = np.argmin(distances)
        clusters[i] = cluster
    # Storing the old centroid values
    C_old = deepcopy(C)
    # Finding the new centroids by taking the average value
    for i in range(k):
        points = [X[j] for j in range(len(X)) if clusters[j] == i]
        C[i] = np.mean(points, axis=0)
    error = dist(C, C_old, None)

colors = ['r', 'g', 'b', 'y', 'c', 'm']

fig, ax = plt.subplots()

for i in range(k):
        points = np.array([X[j] for j in range(len(X)) if clusters[j] == i])
        ax.scatter(points[:, 0], points[:, 1], s=7, c=colors[i])
        
ax.scatter(C[:, 0], C[:, 1], marker='*', s=200, c='#050505')
plt.xlabel('Matemática - G3')
plt.ylabel('Português - G3')
plt.title('Fig. 36 - Agrupamento Da Amostra por Nota de Protuguês vs Matemática (K-means)')
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0)
plt.savefig("agrupamentoKmeans.png")
```
